```python
import pandas as pd
import numpy as np

# Load processed files
balls = pd.read_parquet(r"C:\Users\HP\OneDrive\Desktop\bbl-cricket-dashboard\data\processed\balls.parquet")
info  = pd.read_parquet(r"C:\Users\HP\OneDrive\Desktop\bbl-cricket-dashboard\data\processed\info.parquet")

print(f"Balls shape: {balls.shape}")
print(f"Info shape:  {info.shape}")
```

```python
# See all columns and their data types
print(balls.dtypes)
print("\n")

# Check how many nulls in each column
print("Null counts:")
print(balls.isnull().sum())
```

```python
# Fix date column
balls["start_date"] = pd.to_datetime(balls["start_date"])

# Fix innings to integer
balls["innings"] = balls["innings"].astype(int)

# Fix over number — extract just the over from ball (e.g. 1.3 → over 1)
balls["over"] = balls["ball"].astype(str).str.split(".").str[0].astype(int)

# Fix runs columns — should be numeric
balls["runs_off_bat"] = pd.to_numeric(balls["runs_off_bat"], errors="coerce").fillna(0).astype(int)
balls["extras"]       = pd.to_numeric(balls["extras"], errors="coerce").fillna(0).astype(int)

# Fix all extra type columns
extra_cols = ["wides", "noballs", "legbyes", "byes", "penalty"]
for col in extra_cols:
    balls[col] = pd.to_numeric(balls[col], errors="coerce").fillna(0).astype(int)

print("✅ Data types fixed")
print(balls.dtypes)
```

```python
# Total runs on each ball (batting runs + extras)
balls["total_runs"] = balls["runs_off_bat"] + balls["extras"]

# Is this ball a wicket?
balls["is_wicket"] = balls["wicket_type"].notna().astype(int)

# Is this a legal delivery? (not a wide or no-ball)
balls["is_legal"] = ((balls["wides"] == 0) & (balls["noballs"] == 0)).astype(int)

# Boundary flags
balls["is_four"] = (balls["runs_off_bat"] == 4).astype(int)
balls["is_six"]  = (balls["runs_off_bat"] == 6).astype(int)

# Match phase based on over
def get_phase(over):
    if over < 6:    return "Powerplay"
    elif over < 15: return "Middle"
    else:           return "Death"

balls["phase"] = balls["over"].apply(get_phase)

print("✅ New columns created")
print(balls[["ball", "over", "phase", "runs_off_bat", "extras",
             "total_runs", "is_wicket", "is_four", "is_six"]].head(10))
```

```python
# Check what unique values 'winner' has
winners = info[info["info"] == "winner"][["match_id", "value"]].rename(columns={"value": "winner"})
print(f"Matches with a winner: {len(winners)}")
print(winners.head())
```

```python
# Extract the key info fields we need into a clean match-level table
def extract_info(info_df, category):
    return info_df[info_df["info"] == category][["match_id", "value"]].rename(
        columns={"value": category}
    )

# Pull each category
winner         = extract_info(info, "winner")
toss_winner    = extract_info(info, "toss_winner")
toss_decision  = extract_info(info, "toss_decision")
player_of_match = extract_info(info, "player_of_match")
winner_runs    = extract_info(info, "winner_runs")
winner_wickets = extract_info(info, "winner_wickets")
method         = extract_info(info, "method")   # DLS matches

# Merge all into one clean match-level dataframe
match_info = winner \
    .merge(toss_winner,    on="match_id", how="left") \
    .merge(toss_decision,  on="match_id", how="left") \
    .merge(player_of_match,on="match_id", how="left") \
    .merge(winner_runs,    on="match_id", how="left") \
    .merge(winner_wickets, on="match_id", how="left") \
    .merge(method,         on="match_id", how="left")

print(f"✅ Match info table: {match_info.shape}")
print(match_info.head())
```

```python
# Add winner info to ball-by-ball data
df = balls.merge(match_info, on="match_id", how="left")

print(f"✅ Merged dataset: {df.shape}")
print(df.head())
```

```python
# See what columns you actually have
print(df.columns.tolist())

# Option 1: Only drop columns that exist (safe approach)
drop_cols = [
    "actual_delivery",
    "other_wicket_type",
    "other_player_dismissed",
    "non_boundary",
    "fielder_2",
    "fielder_3"
]
df = df.drop(columns=[col for col in drop_cols if col in df.columns])

# Option 2: Use errors='ignore' (pandas silently skips missing columns)
df = df.drop(columns=drop_cols, errors='ignore')

print(f"✅ Cleaned dataset: {df.shape}")
print(df.columns.tolist())
```

```python
# Check nulls in final dataset
print("Remaining nulls:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# Quick sanity checks
print(f"\nTotal matches:  {df['match_id'].nunique()}")
print(f"Total seasons:  {df['season'].nunique()}")
print(f"Seasons:        {sorted(df['season'].unique())}")
print(f"Total runs:     {df['runs_off_bat'].sum():,}")
print(f"Total wickets:  {df['is_wicket'].sum():,}")
print(f"Total sixes:    {df['is_six'].sum():,}")
print(f"Total fours:    {df['is_four'].sum():,}")
```

```python
df.to_parquet(
    r"C:\Users\HP\OneDrive\Desktop\bbl-cricket-dashboard\data\processed\bbl_final.parquet",
    index=False
)

print("✅ Final clean dataset saved as bbl_final.parquet")
```
