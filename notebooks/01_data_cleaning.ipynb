```python
import pandas as pd
import os
import glob

folder_path = r"C:\Users\HP\OneDrive\Desktop\bbl-cricket-dashboard\data\raw\bbl_male_csv2"

# Get ALL csv files first
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Then separate them cleanly — this works reliably on Windows
ball_files = [f for f in all_files if not f.endswith("_info.csv")]
info_files = [f for f in all_files if f.endswith("_info.csv")]

print(f"Ball files found: {len(ball_files)}")
print(f"Info files found: {len(info_files)}")
```

```python
# Combine ball-by-ball files
ball_dfs = []
for file in ball_files:
    match_id = os.path.basename(file).replace(".csv", "")
    temp_df = pd.read_csv(file, on_bad_lines="skip")  # skips any corrupt rows
    temp_df["match_id"] = match_id
    ball_dfs.append(temp_df)

balls = pd.concat(ball_dfs, ignore_index=True)
print(f"✅ Ball-by-ball data: {balls.shape}")
print(balls.head())
```

```python
# Combine info files
info_dfs = []
for file in info_files:
    match_id = os.path.basename(file).replace("_info.csv", "")
    temp_df = pd.read_csv(file, on_bad_lines="skip")
    temp_df["match_id"] = match_id
    info_dfs.append(temp_df)

info = pd.concat(info_dfs, ignore_index=True)
print(f"✅ Info data: {info.shape}")
print(info.head())
```

```python
info.columns = ["info", "value", "match_id"]

print("✅ Columns renamed")
print(info.head(15))

print("\n📋 All info categories:")
print(info["info"].value_counts())
```

```python
import os

os.makedirs(r"C:\Users\HP\OneDrive\Desktop\bbl-cricket-dashboard\data\processed", exist_ok=True)

balls.to_parquet(r"C:\Users\HP\OneDrive\Desktop\bbl-cricket-dashboard\data\processed\balls.parquet", index=False)
info.to_parquet(r"C:\Users\HP\OneDrive\Desktop\bbl-cricket-dashboard\data\processed\info.parquet", index=False)

print("✅ Both files saved to processed folder!")
```
